```python
import numpy as np
from skimage import io #scipy.misc package 오류나서 대체함
import matplotlib.pyplot as plt

img = io.imread('cake.jpg')
img_tinted = img * [1, 0.8, 0.8]

# Show the original image
plt.subplot(1, 2, 1)
plt.imshow(img)

# Show the tinted image
plt.subplot(1, 2, 2)

# A slight gotcha with imshow is that it might give strange results if presented with 
# data that is not uint8. To work around this, we explicitly cast the image to uint8
# before displaying it.
plt.imshow(np.uint8(img_tinted))
plt.show()
```
