### Exercise 2

Make a square signal at 1500 Hz and make a wave that samples it
at 10000 frames per second.  If you plot the spectrum, you can
see that most of the harmonics are aliased.
When you listen to the wave, can you hear the aliased harmonics? 

### Solution


```python
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import numpy as np

from thinkdsp import SquareSignal
from thinkdsp import decorate
```

1500Hz의 signal을 10000번 sapmling한 square wave가 있다.

```python
square = SquareSignal(1500).make_wave(duration=0.5, framerate=10000)
```

이 square wave의 spectrum은 다음과 같다.

```python
square.make_spectrum().plot()
decorate(xlabel='Frequency (Hz)')
decorate(ylabel='Amplitude')
```

- Fundamental frequency : **1500Hz**
- 첫번째 Harmonic : **4500Hz**

두번째 Harmonic이 7500Hz로 예상되는 것과 달리 **2500Hz**으로 aliasing되었다.

세번째 Harmonic의 경우 10500Hz가 아닌, **-500Hz**

네번째 Harmonic의 경우 13500Hz가 아닌, **3500Hz**

다섯번째 Harmonic의 경우 16500Hz가 아닌, **1500Hz** (Fundamental frequency로 표현됨)

나머지 Harmonic들 또한 이미 존재하는 다른 Harmonic에 overmapping되었다. (Fundamental pitch는 500Hz에 overlap됨)
그렇기 때문에 wave audio를 들을 때, aliasing된 wave때문에 방해되는 일은 없다.

즉, aliasing wave가 발생해도 Harmonic frequency에 overlap되기 때문에 문제되지 않는다.

```python
square.make_audio()
```
